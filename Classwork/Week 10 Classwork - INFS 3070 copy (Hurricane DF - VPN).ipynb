```python
import pandas as pd

import matplotlib.pyplot as plt

# df = pd.read_csv('https://www.nhc.noaa.gov/data/hurdat/hurdat2-1851-2023-051124.txt')

df = pd.read_csv('datafiles\hurricaneData.txt')



df.head()
```

```python
# Define column names while impprting a data file

column_names = [
    'ID-Dates', 'Time', 'RecordID', 'Status', 'Latitude', 'Longitude', 'WindSpeed', 'Pressure', 'ExtraField1', 'Extrafield2',
    'ExtraField3', 'ExtraField4', 'ExtraField5', 'ExtraField6', 'ExtraField7', 'ExtraField8', 'ExtraField9', 'ExtraField10', 
    'ExtraField11', 'ExtraField12'
]

# Read the file, restrict it to the number of columns and column names we want to use

df = pd.read_csv('datafiles\hurricaneData.txt', names = column_names, usecols = range(len(column_names)), on_bad_lines = 'skip')

df.head()
```

### Extract ID, Year, and Month form ID-Dates Column

```python
# Create new columns in our dataframe with empty values for year and month

df['YEAR'] = ''
df['MONTH'] = ''

# Only extract the Year and Month where 'ID-Dates' does not contain 'AL'
# Extract the year

df.loc[~df['ID-Dates'].str.contains('AL'), 'YEAR'] = df['ID-Dates'].str[:4]

# Extract month

df.loc[~df['ID-Dates'].str.contains('AL'), 'MONTH'] = df['ID-Dates'].str[4:6]

# Create a unique ID column that copies the AL identifier

df['UNIQUE_ID'] = df['ID-Dates'].apply(lambda x: x if 'AL' in x else None)

# To copy this to the entire dataframe we're going to use forward 'Fordward-Fill'"Week 10 Classwork - INFS 3070.ipynb"
df['UNIQUE_ID'] = df['UNIQUE_ID'].ffill()

df.head(5)
```

### Number of hurricanes by year

```python
hurricanes_by_year = df.groupby('YEAR')['UNIQUE_ID'].count().reset_index(name = 'COUNT')

hurricanes_by_year
```

```python
# Visualize the number of storms

# Convert thye year column to numeric, coerce errors to Null

df['YEAR'] = pd.to_numeric(df['YEAR'], errors = 'coerce')

# Drop rows where YEAR is NaN (this is the unique id rows)

df_clean = df.dropna(subset = ['YEAR'])

# Convert year to an integer using .loc
# The [:, 'YEAR'] instructs Pandas to select all rows in the 'YEAR' column

df_clean.loc[:, 'YEAR'] = df_clean['YEAR'].astype(int)

# Now group the cleansed data

hurricanes_by_year = df_clean.groupby('YEAR')['UNIQUE_ID'].nunique()

# Visualize

plt.figure(figsize = (10,5))
plt.plot(hurricanes_by_year.index, hurricanes_by_year.values, marker = 'o')

plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.title('Hurricanes by Year')


# Controls interval of the X axis values
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(25))
plt.xticks(rotation = 45)
plt.grid(True)

plt.show()
```

```python
# Visualize the number of storms

# Convert thye year column to numeric, coerce errors to Null

df['YEAR'] = pd.to_numeric(df['YEAR'], errors = 'coerce')

# Drop rows where YEAR is NaN (this is the unique id rows)

df_clean = df.dropna(subset = ['YEAR'])

# Convert year to an integer using .loc
# The [:, 'YEAR'] instructs Pandas to select all rows in the 'YEAR' column

df_clean.loc[:, 'YEAR'] = df_clean['YEAR'].astype(int)

# Now group the cleansed data

hurricanes_by_year = df_clean.groupby('YEAR')['UNIQUE_ID'].nunique()

# Calculate the cumulative average of hurricanes

cummulative_avg_hurricanes = hurricanes_by_year.expanding().mean()

# Visualize

plt.figure(figsize = (10,5))
plt.plot(hurricanes_by_year.index, hurricanes_by_year.values, marker = 'o')

# Plot an average line

plt.plot(cummulative_avg_hurricanes.index, cummulative_avg_hurricanes.values, color = 'red', linestyle = '--', label = 'Cummulative Average')

plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.title('Hurricanes by Year')


# Controls interval of the X axis values
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(25))
plt.xticks(rotation = 45)
plt.grid(True)

plt.show()
```

### Hurricanes by category

Here we introduce the 'Functions' within python

```python
# Create a functuon to define catergories based on wind speed

def categorize_wind_speed(speed):
    if 74 <= speed <= 95:
        return "Category 1"
    elif 96 <= speed <= 110:
        return "Category 2"
    elif 111 <= speed <= 129:
        return "Category 3"
    elif 130 <= speed <= 156:
        return "Category 4"
    elif speed >= 157:
        return "Category 5"


# Create a new column in our dataframe called 'CATEGORY' and populate it with teh Category based on the windspeed
df['CATEGORY'] = df['WindSpeed'].apply(categorize_wind_speed)

# Group our data

hurricanes_by_year_category = df.groupby(['YEAR', 'CATEGORY'])['UNIQUE_ID'].count().unstack(fill_value = 0)

# Plot the graph

plt.figure(figsize = (12,6))

# Loop through the data to read every row

for category in hurricanes_by_year_category.columns:
    plt.plot(hurricanes_by_year_category.index, hurricanes_by_year_category[category], label = 'category')

plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.title('Number of Hurricanes per Year and Category')
plt.grid()
plt.legend()
plt.show()
```
