### What is the name and year of the movie with the highest rating and highest number of votes?

This will require us to learn the merge function to merge multiple data sets together

```python
# RE-import dfTitles to get all of the columns back
dfTitles = pd.read_csv('http://128.198.162.191/pydev/dataFiles/IMDB/titleBasics.tsv', sep='\t', low_memory = False)

# We need to merge Titles and Ratings

mvRating = pd.merge(dfTitles, dfRatings, on = 'tconst')

mvRating
```

### Query data frame to reduce the amount of data

```python
# Only return the records for the type of 'movie'

ratingSort = (mvRating.query('titleType == "movie"'))

ratingSort
```

### Sort our dataFrame by ratings

```python
# Return a list of movies and their titles sorted by 'averageRating'

ratingSort.sort_values(by = ['averageRating', 'numVotes'], ascending = False)
```

```python
ratingSort.sort_values(by = ['numVotes', 'averageRating'], ascending = False)
```

### Format the results for distribution

```python
# Create a dataFrame with the queried and sorted data

semi1 = (ratingSort.sort_values(by = ['numVotes', 'averageRating'], ascending = False))

# Create a dataFrame with only the required information

semi2 = semi1[['primaryTitle', 'startYear', 'averageRating', 'numVotes']]

# Select a subset of the columns for a summary

final = (semi2.head(10))

print(final)

```

### Output results to a file

```python
final.to_csv('datafiles.csv', index = False)
```
